```python
import pandas as pd
import numpy as np
from scipy.spatial import ConvexHull
import matplotlib.pyplot as plt
import geopandas as gpd
from shapely.geometry import Polygon
from sklearn.cluster import KMeans
from shapely.geometry.multipolygon import MultiPolygon
```

```python
df = pd.read_excel(r"C:/Users/admin/Desktop/hug inn/ddata.xlsx")
```

```python
k=4
kmean=KMeans(n_clusters=k)
kmean = kmean.fit(df)
y = kmean.fit_predict(df[['Long', 'Lat', 'EC 0-2']])
df['Cluster'] = y
```

```python
df['points']=df[['Long', 'Lat','Cluster']].apply(tuple, axis=1)
points=df['points']

c0 = []
c1 = []
c2 = []
c3 = []

for line in points:
    s = []
    s.append(line[0])
    s.append(line[1])
    if line[2] == 0.0:
        c0.append(s)
    if line[2] == 1.0:
        c1.append(s)
    if line[2] == 2.0:
        c2.append(s)
    if line[2] == 3.0:
        c3.append(s)

hull0 = ConvexHull(c0)
hull1 = ConvexHull(c1)
hull2 = ConvexHull(c2)
hull3 = ConvexHull(c3)

c0 = np.array(c0)
c1 = np.array(c1)
c2 = np.array(c2)
c3 = np.array(c3)
```

```python
bp0=[]
bp1=[]
bp2=[]
bp3=[]
```

```python
plt.plot(c0[:,0], c0[:,1], 'o')
for simplex in hull0.simplices:
    bp0.append(c0[simplex[0]])
    bp0.append(c0[simplex[1]])
    plt.plot(c0[simplex,0], c0[simplex,1], 'k-') 

plt.show()
```

```python
plt.plot(c1[:,0], c1[:,1], 'o')
for simplex in hull1.simplices:
    bp1.append(c1[simplex[0]])
    bp1.append(c1[simplex[1]])
    plt.plot(c1[simplex,0], c1[simplex,1], 'k-') 

plt.show()
```

```python
plt.plot(c2[:,0], c2[:,1], 'o')
for simplex in hull2.simplices:
    bp2.append(c2[simplex[0]])
    bp2.append(c2[simplex[1]])
    plt.plot(c2[simplex,0], c2[simplex,1], 'k-') 

plt.show()
```

```python
plt.plot(c3[:,0], c3[:,1], 'o')
for simplex in hull3.simplices:
    bp3.append(c3[simplex[0]])
    bp3.append(c3[simplex[1]])
    plt.plot(c3[simplex,0], c3[simplex,1], 'k-') 

plt.show()
```

```python
from shapely.geometry.multipolygon import MultiPolygon
#from shapely.ops import cascaded_union

crs = {'init': 'epsg:4326'}

#poly_geometry= Polygon(bp0)
poly = gpd.GeoDataFrame( crs=crs,geometry=[ Polygon(bp0),Polygon(bp1),Polygon(bp2),Polygon(bp3) ])       

poly= poly.convex_hull


#poly = gpd.overlay(poly0, poly1,how='intersection')

#print(poly.geometry)

poly.to_file(filename='polygon.shp', driver="ESRI Shapefile")
```

```python
from osgeo import ogr
driver = ogr.GetDriverByName('ESRI Shapefile')
dataSource = driver.Open('C:/Users/admin/Desktop/hug inn/polygon.shp', 1) #1 is read/write

#define floating point field named DistFld and 16-character string field named Name:
fldDef = ogr.FieldDefn('DistFld', ogr.OFTReal)
fldDef2 = ogr.FieldDefn('color', ogr.OFTString)
fldDef2.SetWidth(16) #16 char string width

#get layer and add the 2 fields:
layer = dataSource.GetLayer()
layer.CreateField(fldDef)
layer.CreateField(fldDef2)
```
