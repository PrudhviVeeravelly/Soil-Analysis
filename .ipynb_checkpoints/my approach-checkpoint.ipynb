```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()  # for plot styling
import numpy as np
import pandas as pd
```

```python
from sklearn.cluster import KMeans
```

```python
df = pd.read_excel(r"C:/Users/admin/Desktop/hug inn/ddata.xlsx")
df
```

```python
k=4
kmean=KMeans(n_clusters=4)
```

```python
kmean = kmean.fit(df)
```

```python
y = kmean.fit_predict(df[['Long', 'Lat', 'EC 0-2']])
df['Cluster'] = y
df
```

```python
labels=kmean.labels_  

set(labels)
```

```python
colors=['blue','black','green','red']
```

```python
from mpl_toolkits import mplot3d
```

```python
plt.axes(projection = '3d')
y=0
for x in labels:
    plt.scatter(df.iloc[y,0],df.iloc[y,1],df.iloc[y,2],color=colors[x])   
    y=y+1

    
plt.xlabel('lat')
plt.ylabel('long')
plt.title(' Number of clusters k= {} '.format(k) )


plt.show()
```
